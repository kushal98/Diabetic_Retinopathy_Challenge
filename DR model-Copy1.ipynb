```python
from PIL import Image
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
```

```python
rows,cols = 224,224
```

```python
immatrix = []
imlabels = []
```

```python
for file in os.listdir("./resized/"):
    base = os.path.basename("./resized/" + file)
    print(file)
    fileName = os.path.splitext(base)[0]
    imlabels.append(train_labels.loc[train_labels.image==fileName, 'level'].values[0])
    im = Image.open("./resized/" + file)
    img = im.resize((rows,cols))
    immatrix.append(np.array(img))
```

```python
immatrix = np.asarray(immatrix)
imlabels = np.asarray(imlabels)
```

```python
immatrix[0].shape
```

```python
imlabels[1].shape
```

```python
from sklearn.utils import shuffle

data,Label = shuffle(immatrix,imlabels, random_state=2)
train_data = [data,Label]
type(train_data)
```

```python
(X,y) = (train_data[0],train_data[1])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

print(X_train.shape)
print(X_test.shape)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

#Normalizing Data Between [0,1]
X_train /= 255
X_test /= 255

print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
y_train
```

```python
X_train[0].shape
```

```python
from keras.utils import np_utils

n_classes = 5

Y_train = np_utils.to_categorical(y_train, n_classes)
Y_test = np_utils.to_categorical(y_test, n_classes)
```

```python
Y_train[0]
```

```python
from keras.applications.vgg16 import VGG16
base_model = VGG16(weights='imagenet',include_top=False, input_shape=(224,224,3))
print(base_model.summary())
```

```python
from keras.models import Sequential
from keras.layers import Dense , Flatten , Dropout, Activation

top_model = Sequential()
top_model.add(Flatten(input_shape=base_model.output_shape[1:]))
top_model.add(Dense(256, activation='relu'))
top_model.add(Dropout(0.5))
top_model.add(Dense(5, activation='softmax'))
```

```python
from keras.models import Model

model=Model(inputs=base_model.input, outputs=top_model(base_model.output))
```

```python
from keras import optimizers

model.compile(loss='categorical_crossentropy',
     optimizer=optimizers.SGD(lr=1e-4, momentum=0.9),
     metrics=['accuracy'])
model.summary()
```

```python
batch_size =32
n_epochs =20
n_batches = 1000//batch_size
```

```python
with tf.device("/gpu:0"):
    model.fit(x = X_train,y=Y_train)
```

```python
with tf.device("/gpu:0"):
    probs = model.predict(X_test)
```
